```python
import csv
import numpy as np
```

```python
list=['hurto', 'atraco', 'chorear', 'raponazo', 'cosquilleo', 'bajar', 'tumbar', 'jalar', 'ladron','rata','fleteo','quitaron','chalequear','arrancar','desvalijar']
archivo=open('Data Twitter/TWITTER_COL/TWITTER_COLOMBIA_FREQ.csv')
reader=csv.reader(archivo)
```

```python
palabras=[]
for index, item in enumerate(reader):
    if index==0:
        ciudad=item[1:]
    if index==5:
        total=item[6:]    
    if index >5 and item[0] in list:
        palabras.append(item)
    
```

```python
n=len(palabras[2])-1
suma=[0]*n
```

```python
for p in palabras:
    for i in range(n):
        suma[i]+=int(p[i+1])
for i in range(len(total)):
    suma[i+5]=suma[i+5]*1.0/int(total[i])
```

```python
import matplotlib.pyplot as plt
c=ciudad[5:80]
plt.figure(figsize=(120,5))
plt.bar(c,suma[5:80])
plt.show()
```

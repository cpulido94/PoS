```python
# parametros modelo
modelo='g_m_v'
T=300
psi=0.98
nu=0.8
mu=0.14
lamda={'A':0,'B':0.01,'C':0.1}
```

```python
s0=np.random.rand(len(vertices))
```

```python
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(331,ylim=(0,1))
ax2 = fig.add_subplot(334,ylim=(0,1))
colors=sns.color_palette()[:2]
#########################
vertices,edges=read("poor_connected.graph")
grado_grupo =np.array([[len(vertice[2:]),grupos[vertice[1]]] for vertice in vertices])            
V=pd.DataFrame(grado_grupo,columns=['Grade','Group']).astype({'Grade': 'int64'})
G=convert_vertices_to_graph(vertices,edges)

S,conteo=generate(vertices,psi=psi,
           nu=nu,mu=mu,T=T,
           s=s0,lamda=lamda,modelo=modelo,contar=True)
S=S.T
V['conections']=conteo

# indices de cada grupo con maximo grado, minimo grado y uno aleatorio
index_examples=[]
for g in grupos.values():
    index_examples.append([V[V.Group == g].Grade.idxmax(),V[V.Group == g].Grade.idxmin()])
    
X=pd.DataFrame(index=np.arange(T*6))
X['Fear']=np.array([list(S[index_examples[i]].flatten()) for i in range(len(index_examples))]).flatten()
X['Time']=list(np.arange(T))*6
X['Group']=np.array([[list(grupos.values())[i]]*2*T for i in range(len(grupos))]).flatten()
X['Grade nodes']=(['max']*T+['min']*T)*3  

A=pd.DataFrame()
A['Fear']=np.concatenate((S[V.Group == 'Immune'].mean(axis=0),
                S[V.Group == 'Susceptible'].mean(axis=0),
                S[V.Group == 'Highly Susceptible'].mean(axis=0)))
A['Group']=['Immune']*T+['Susceptible']*T+['Highly Susceptible']*T

sns.boxenplot(data=X,x='Group',y='Fear',hue='Grade nodes',palette=colors,ax=ax1,order=grupos.values())
ax1.legend(loc='upper center',ncol=4,bbox_to_anchor=(0.45,1.23))
ax1.set(title="Poor Connected")
sns.violinplot(data=A,y='Fear',x='Group',palette=['g','b','r'],ax=ax2,order=grupos.values())
##########################

ax1 = fig.add_subplot(332,ylim=(0,1))
ax2 = fig.add_subplot(335,ylim=(0,1))
colors=sns.color_palette()[:2]
vertices,edges=read("medium_connected.graph")
grado_grupo =np.array([[len(vertice[2:]),grupos[vertice[1]]] for vertice in vertices])            
V=pd.DataFrame(grado_grupo,columns=['Grade','Group']).astype({'Grade': 'int64'})
G=convert_vertices_to_graph(vertices,edges)

S,conteo=generate(vertices,psi=psi,
           nu=nu,mu=mu,T=T,
           s=s0,lamda=lamda,modelo=modelo,contar=True)
S=S.T
V['conections']=conteo

# indices de cada grupo con maximo grado, minimo grado y uno aleatorio
index_examples=[]
for g in grupos.values():
    index_examples.append([V[V.Group == g].Grade.idxmax(),V[V.Group == g].Grade.idxmin()])
    
X=pd.DataFrame(index=np.arange(T*6))
X['Fear']=np.array([list(S[index_examples[i]].flatten()) for i in range(len(index_examples))]).flatten()
X['Time']=list(np.arange(T))*6
X['Group']=np.array([[list(grupos.values())[i]]*2*T for i in range(len(grupos))]).flatten()
X['Grade nodes']=(['max']*T+['min']*T)*3  

A=pd.DataFrame()
A['Fear']=np.concatenate((S[V.Group == 'Immune'].mean(axis=0),
                S[V.Group == 'Susceptible'].mean(axis=0),
                S[V.Group == 'Highly Susceptible'].mean(axis=0)))
A['Group']=['Immune']*T+['Susceptible']*T+['Highly Susceptible']*T

sns.boxenplot(data=X,x='Group',y='Fear',hue='Grade nodes',palette=colors,ax=ax1,order=grupos.values())
ax1.legend(loc='upper center',ncol=4,bbox_to_anchor=(0.45,1.23))
ax1.set(title="Medium Connected")
sns.violinplot(data=A,y='Fear',x='Group',palette=['g','b','r'],ax=ax2,order=grupos.values())
#########################

ax1 = fig.add_subplot(333,ylim=(0,1))
ax2 = fig.add_subplot(336,ylim=(0,1))
colors=sns.color_palette()[:2]
vertices,edges=read("highly_connected.graph")
grado_grupo =np.array([[len(vertice[2:]),grupos[vertice[1]]] for vertice in vertices])            
V=pd.DataFrame(grado_grupo,columns=['Grade','Group']).astype({'Grade': 'int64'})
G=convert_vertices_to_graph(vertices,edges)

S,conteo=generate(vertices,psi=psi,
           nu=nu,mu=mu,T=T,
           s=s0,lamda=lamda,modelo=modelo,contar=True)
S=S.T
V['conections']=conteo

# indices de cada grupo con maximo grado, minimo grado y uno aleatorio
index_examples=[]
for g in grupos.values():
    index_examples.append([V[V.Group == g].Grade.idxmax(),V[V.Group == g].Grade.idxmin()])
    
X=pd.DataFrame(index=np.arange(T*6))
X['Fear']=np.array([list(S[index_examples[i]].flatten()) for i in range(len(index_examples))]).flatten()
X['Time']=list(np.arange(T))*6
X['Group']=np.array([[list(grupos.values())[i]]*2*T for i in range(len(grupos))]).flatten()
X['Grade nodes']=(['max']*T+['min']*T)*3  

A=pd.DataFrame()
A['Fear']=np.concatenate((S[V.Group == 'Immune'].mean(axis=0),
                S[V.Group == 'Susceptible'].mean(axis=0),
                S[V.Group == 'Highly Susceptible'].mean(axis=0)))
A['Group']=['Immune']*T+['Susceptible']*T+['Highly Susceptible']*T

sns.boxenplot(data=X,x='Group',y='Fear',hue='Grade nodes',palette=colors,ax=ax1,order=grupos.values())
ax1.legend(loc='upper center',ncol=4,bbox_to_anchor=(0.45,1.23))
ax1.set(title="Highly Connected")
sns.violinplot(data=A,y='Fear',x='Group',palette=['g','b','r'],ax=ax2,order=grupos.values())
plt.show()
```
